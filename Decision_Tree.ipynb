```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
# Load iris dataset
iris = pd.read_csv("iris (1).csv")
```

```python
# Separate features and target
X_iris = iris.iloc[:, :-1]   # all columns except last
y_iris = iris.iloc[:, -1]    # last column (target)
```

```python
# Split dataset (80% train, 20% test)
X_train_i, X_test_i, y_train_i, y_test_i = train_test_split(
    X_iris, y_iris, test_size=0.2, random_state=42
)
```

```python
# Create Decision Tree model
dt_iris = DecisionTreeClassifier()
```

```python
# Train the model
dt_iris.fit(X_train_i, y_train_i)

# Predict on test data
y_pred_i = dt_iris.predict(X_test_i)

# Evaluate model
print("IRIS DATASET RESULTS")
print("Accuracy:", accuracy_score(y_test_i, y_pred_i))
print("Confusion Matrix:\n", confusion_matrix(y_test_i, y_pred_i))
```

```python
# -----------------------------
# 2. Decision Tree for Drug Dataset
# -----------------------------
```

```python
# Load drug dataset
drug = pd.read_csv("drug.csv")

# Convert categorical features to numeric using one-hot encoding
drug_encoded = pd.get_dummies(drug, drop_first=True)

# Separate features and target
X_drug = drug_encoded.drop("Drug_drugY", axis=1) if "Drug_drugY" in drug_encoded.columns else drug_encoded.iloc[:, :-1]
y_drug = drug.iloc[:, -1]
```

```python
# Split dataset
X_train_d, X_test_d, y_train_d, y_test_d = train_test_split(
    X_drug, y_drug, test_size=0.2, random_state=42
)

# Create Decision Tree model
dt_drug = DecisionTreeClassifier()

# Train the model
dt_drug.fit(X_train_d, y_train_d)

# Predict on test data
y_pred_d = dt_drug.predict(X_test_d)

# Evaluate model
print("\nDRUG DATASET RESULTS")
print("Accuracy:", accuracy_score(y_test_d, y_pred_d))
print("Confusion Matrix:\n", confusion_matrix(y_test_d, y_pred_d))
```
