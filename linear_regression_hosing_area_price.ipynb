```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
# Load dataset
df = pd.read_csv("canada_per_capita_income.csv")

# Display first 5 rows
df.head()
```

```python
# Independent variable (Year)
X = df[['year']]

# Dependent variable (Income)
y = df['per capita income (US$)']
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
slope = model.coef_[0]
intercept = model.intercept_

print("Slope (m):", slope)
print("Intercept (b):", intercept)
```

```python
predicted_income = model.predict([[2020]])

print("Predicted Per Capita Income for 2020: $", predicted_income[0])
```

```python
y_pred = model.predict(X)
r2 = r2_score(y, y_pred)

print("R2 Score:", r2)
```

```python
plt.figure()
plt.scatter(X, y)
plt.plot(X, y_pred)
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.title("Canada Per Capita Income Prediction")
plt.show()
```

```python
# Open and read file
file = open("canada_per_capita_income.csv", "r")

years = []
income = []

# Skip header
next(file)

for line in file:
    row = line.strip().split(",")
    years.append(float(row[0]))
    income.append(float(row[1]))

file.close()

print("Years:", years)
print("Income:", income)
```

```python
N = len(years)

sum_x = sum(years)
sum_y = sum(income)

sum_xy = 0
sum_x2 = 0

for i in range(N):
    sum_xy += years[i] * income[i]
    sum_x2 += years[i] * years[i]

# Calculate slope (m)
m = (N * sum_xy - sum_x * sum_y) / (N * sum_x2 - sum_x * sum_x)

# Calculate intercept (b)
b = (sum_y - m * sum_x) / N

print("Slope (m):", m)
print("Intercept (b):", b)
```

```python
year_2020 = 2020
predicted_income = m * year_2020 + b

print("Predicted Per Capita Income for 2020:", predicted_income)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv("salary.csv")

# Display first 5 rows
df.head()
```

```python
X = df[['YearsExperience']]
y = df['Salary']
```

```python
model = LinearRegression()

# Find rows where X has NaN values
nan_rows = X.isnull().any(axis=1)

# Drop corresponding rows from X and y
X_cleaned = X[~nan_rows]
y_cleaned = y[~nan_rows]

model.fit(X_cleaned, y_cleaned)
```

```python
slope = model.coef_[0]
intercept = model.intercept_

print("Slope (m):", slope)
print("Intercept (b):", intercept)

print("\nRegression Equation:")
print("Salary =", slope, "* YearsExperience +", intercept)
```

```python
predicted_salary = model.predict([[12]])

print("Predicted Salary for 12 years experience: $", predicted_salary[0])
```

```python
y_pred_cleaned = model.predict(X_cleaned)
r2 = r2_score(y_cleaned, y_pred_cleaned)

print("R2 Score:", r2)
```

```python
plt.figure()
plt.scatter(X_cleaned, y_cleaned) # Use cleaned data for scatter plot
plt.plot(X_cleaned, y_pred_cleaned, color='red') # Use cleaned data and its corresponding predictions for the regression line
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Salary Prediction using Linear Regression")
plt.show()
```

```python
import csv

# Read data from CSV
x = []
y = []

with open("salary.csv", "r") as file:
    data = csv.reader(file)
    next(data)   # Skip header

    for row in data:
        # Skip row if either value is empty
        if row[0] and row[1]:
            x.append(float(row[0]))   # YearsExperience
            y.append(float(row[1]))   # Salary

# Number of observations
n = len(x)

# Calculate mean of x and y
mean_x = sum(x) / n
mean_y = sum(y) / n

# Calculate slope (b1)
numerator = 0
denominator = 0

for i in range(n):
    numerator += (x[i] - mean_x) * (y[i] - mean_y)
    denominator += (x[i] - mean_x) ** 2

b1 = numerator / denominator

# Calculate intercept (b0)
b0 = mean_y - b1 * mean_x

print("Slope (b1):", b1)
print("Intercept (b0):", b0)

# Predict salary for 12 years experience
years = 12
predicted_salary = b0 + b1 * years

print("\nPredicted Salary for 12 years experience:", predicted_salary)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("canada_per_capita_income.csv")

# Check missing values
print(df.isnull().sum())

# Define features and target
X = df[['year']]
y = df['per capita income (US$)']

# Create model
model = LinearRegression()
model.fit(X, y)

# Predict income for 2020
pred_2020 = model.predict([[2020]])
print("Predicted Per Capita Income in 2020:", pred_2020[0])

# Plot data and regression line
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("Year")
plt.ylabel("Per Capita Income")
plt.title("Year vs Per Capita Income")
plt.show()
```
